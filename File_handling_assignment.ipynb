```python
2.What is exception handling in Python ?
Exception handling lets you detect, handle, and respond to errors without stopping the program.
Python uses the following blocks:
1. try:
The code that may cause an error goes here.
2. except:
The code that runs when an error occurs.
3. else (optional):
Runs only if no error occurs.
4. finally (optional):
Runs always, whether an error occurs or not.
```

```python
15.How does Python's garbage collection system work?

Python uses a combination of:
1.Reference Counting (Primary method)
2.Garbage Collector for Cyclic Garbage (Secondary method)
1. Reference Counting (Main Mechanism)
Every Python object keeps track of how many references (variables or objects) are pointing to it.
When a new reference is created → reference count increases
When a reference is deleted → reference count decreases
When the reference count becomes 0, the object is automatically deleted
2. Garbage Collector (Handles Circular References)
Sometimes objects refer to each other, creating a cycle.
```

```python
6.Write a Python program that logs an error message to a log file when a division by zero exception occurs.
import logging

logging.basicConfig(
    filename="error.log",
    level=logging.ERROR,
    format="%(asctime)s - %(levelname)s - %(message)s"
)

try:
    a = 10
    b = 0
    result = a / b
    print(result)

except ZeroDivisionError as e:
    logging.error("Division by zero attempted: %s", e)
    print("An error occurred. Check the log file for details.")
```

```python
16.Demonstrate how to use memory profiling to check the memory usage of a small program.
import os
import psutil

def memory_usage():
    process = psutil.Process(os.getpid())
    return process.memory_info().rss / (1024 ** 2)

before = memory_usage()

lst = [i for i in range(200000)]

after = memory_usage()

print(f"Memory before: {before:.2f} MB")
print(f"Memory after: {after:.2f} MB")
print(f"Memory increased by: {after - before:.2f} MB")
```

```python
18.How would you implement a basic logging setup that logs to a file with rotation after 1MB.
import logging
from logging.handlers import RotatingFileHandler

handler = RotatingFileHandler(
    "app.log",
    maxBytes=1_000_000,
    backupCount=5
)

formatter = logging.Formatter("%(asctime)s - %(levelname)s - %(message)s")
handler.setFormatter(formatter)
logger = logging.getLogger("my_logger")
logger.setLevel(logging.INFO)
logger.addHandler(handler)

for i in range(10000):
    logger.info(f"Log entry number {i}")
```
